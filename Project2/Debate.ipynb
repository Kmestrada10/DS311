```python
def compare_basic_stats(messi_df, ronaldo_df):
    # Total goals
    messi_goals = messi_df['goals'].sum()
    ronaldo_goals = ronaldo_df['goals'].sum()
    
    # Total assists
    messi_assists = messi_df['assists'].sum()
    ronaldo_assists = ronaldo_df['assists'].sum()
    
    # Total matches
    messi_matches = len(messi_df)
    ronaldo_matches = len(ronaldo_df)
    
    # Goals per match
    messi_gpm = messi_goals / messi_matches
    ronaldo_gpm = ronaldo_goals / ronaldo_matches
    
    # Assists per match
    messi_apm = messi_assists / messi_matches
    ronaldo_apm = ronaldo_assists / ronaldo_matches
    
    # Goal contributions (goals + assists) per match
    messi_gcpm = (messi_goals + messi_assists) / messi_matches
    ronaldo_gcpm = (ronaldo_goals + ronaldo_assists) / ronaldo_matches
    
    # Create comparison DataFrame
    comparison = pd.DataFrame({
        'Statistic': ['Total Goals', 'Total Assists', 'Total Matches', 
                     'Goals per Match', 'Assists per Match', 'Goal Contributions per Match'],
        'Messi': [messi_goals, messi_assists, messi_matches, 
                 messi_gpm, messi_apm, messi_gcpm],
        'Ronaldo': [ronaldo_goals, ronaldo_assists, ronaldo_matches,
                   ronaldo_gpm, ronaldo_apm, ronaldo_gcpm]
    })
    
    return comparison

# Now call the function with your actual DataFrames
basic_stats = compare_basic_stats(messi_df, ronaldo_df)
basic_stats
```

```python
def compare_by_competition(messi_df, ronaldo_df):
    # Messi by competition
    messi_comp = messi_df.groupby('competition').agg({
        'goals': 'sum',
        'assists': 'sum',
        'match_no': 'count'
    }).rename(columns={'match_no': 'matches'})
    
    # Ronaldo by competition
    ronaldo_comp = ronaldo_df.groupby('competition').agg({
        'goals': 'sum',
        'assists': 'sum',
        'match_no': 'count'
    }).rename(columns={'match_no': 'matches'})
    
    # Calculate goals per match
    messi_comp['gpm'] = messi_comp['goals'] / messi_comp['matches']
    ronaldo_comp['gpm'] = ronaldo_comp['goals'] / ronaldo_comp['matches']
    
    return messi_comp, ronaldo_comp

messi_by_comp, ronaldo_by_comp = compare_by_competition(messi_df, ronaldo_df)
print("Messi by Competition:")
display(messi_by_comp)
print("\nRonaldo by Competition:")
display(ronaldo_by_comp)
```

```python
def compare_big_matches(messi_df, ronaldo_df):
    # Filter for big matches
    messi_big = messi_df[messi_df['stage'].notna()]
    ronaldo_big = ronaldo_df[ronaldo_df['stage'].notna()]
    
    big_stats = pd.DataFrame({
        'Player': ['Messi', 'Ronaldo'],
        'Big Matches': [len(messi_big), len(ronaldo_big)],
        'Goals': [messi_big['goals'].sum(), ronaldo_big['goals'].sum()],
        'Assists': [messi_big['assists'].sum(), ronaldo_big['assists'].sum()],
        'Goals per Match': [messi_big['goals'].sum()/len(messi_big), 
                         ronaldo_big['goals'].sum()/len(ronaldo_big)],
        'Assists per Match': [messi_big['assists'].sum()/len(messi_big),
                            ronaldo_big['assists'].sum()/len(ronaldo_big)]
    })
    return big_stats

big_match_stats = compare_big_matches(messi_df, ronaldo_df)
big_match_stats
```

```python
def compare_shooting(messi_df, ronaldo_df):
    messi_shots = messi_df[messi_df['shots'].notna()]
    ronaldo_shots = ronaldo_df[ronaldo_df['shots'].notna()]
    
    messi_conversion = (messi_shots['shots_on_target'].sum() / messi_shots['shots'].sum()) * 100
    ronaldo_conversion = (ronaldo_shots['shots_on_target'].sum() / ronaldo_shots['shots'].sum()) * 100
    
    shooting_stats = pd.DataFrame({
        'Statistic': ['Total Shots', 'Shots on Target', 'Conversion Rate (%)'],
        'Messi': [messi_shots['shots'].sum(), 
                 messi_shots['shots_on_target'].sum(),
                 messi_conversion],
        'Ronaldo': [ronaldo_shots['shots'].sum(),
                   ronaldo_shots['shots_on_target'].sum(),
                   ronaldo_conversion]
    })
    return shooting_stats

shooting_comparison = compare_shooting(messi_df, ronaldo_df)
shooting_comparison
```

```python
def compare_motm(messi_df, ronaldo_df):
    messi_motm = len(messi_df[messi_df['motm'].notna()])
    ronaldo_motm = len(ronaldo_df[ronaldo_df['motm'].notna()])
    
    messi_motm_rate = messi_motm / len(messi_df)
    ronaldo_motm_rate = ronaldo_motm / len(ronaldo_df)
    
    motm_stats = pd.DataFrame({
        'Statistic': ['Total MOTM Awards', 'MOTM per Match'],
        'Messi': [messi_motm, messi_motm_rate],
        'Ronaldo': [ronaldo_motm, ronaldo_motm_rate]
    })
    return motm_stats

motm_comparison = compare_motm(messi_df, ronaldo_df)
motm_comparison
```

```python
def plot_goals_assists_over_time(messi_df, ronaldo_df):
    messi_df['year'] = pd.to_datetime(messi_df['date']).dt.year
    ronaldo_df['year'] = pd.to_datetime(ronaldo_df['date']).dt.year

    messi_df['year'] = messi_df['year'].astype(int)
    ronaldo_df['year'] = ronaldo_df['year'].astype(int)

    messi_stats = messi_df.groupby('year')[['goals', 'assists']].sum().reset_index()
    ronaldo_stats = ronaldo_df.groupby('year')[['goals', 'assists']].sum().reset_index()

    max_val = max(messi_stats[['goals', 'assists']].values.max(),
                  ronaldo_stats[['goals', 'assists']].values.max())

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

    # Messi
    ax1.plot(messi_stats['year'], messi_stats['goals'], label='Goals', color='dodgerblue', marker='o')
    ax1.plot(messi_stats['year'], messi_stats['assists'], label='Assists', color='skyblue', marker='s')
    ax1.set_title('Messi')
    ax1.set_xlabel('Year')
    ax1.set_ylabel('Count')
    ax1.grid(True)
    ax1.set_ylim(0, max_val + 10)
    ax1.set_xticks(messi_stats['year'])
    ax1.tick_params(axis='x', rotation=45)
    ax1.legend()

    # Ronaldo
    ax2.plot(ronaldo_stats['year'], ronaldo_stats['goals'], label='Goals', color='crimson', marker='o')
    ax2.plot(ronaldo_stats['year'], ronaldo_stats['assists'], label='Assists', color='lightcoral', marker='s')
    ax2.set_title('Ronaldo')
    ax2.set_xlabel('Year')
    ax2.grid(True)
    ax2.set_ylim(0, max_val + 10)
    ax2.set_xticks(ronaldo_stats['year'])
    ax2.tick_params(axis='x', rotation=45)
    ax2.legend()

    plt.suptitle('Yearly Goals & Assists Comparison', fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_goals_assists_over_time_bar(messi_df, ronaldo_df):
    # Convert and group
    messi_df['year'] = pd.to_datetime(messi_df['date']).dt.year.astype(int)
    ronaldo_df['year'] = pd.to_datetime(ronaldo_df['date']).dt.year.astype(int)

    messi_stats = messi_df.groupby('year')[['goals', 'assists']].sum().reset_index()
    ronaldo_stats = ronaldo_df.groupby('year')[['goals', 'assists']].sum().reset_index()

    all_years = sorted(set(messi_stats['year']).union(set(ronaldo_stats['year'])))
    spacing = 2.2
    y_positions = [i * spacing for i in range(len(all_years))]

    messi_stats = messi_stats.set_index('year').reindex(all_years, fill_value=0).reset_index()
    ronaldo_stats = ronaldo_stats.set_index('year').reindex(all_years, fill_value=0).reset_index()

    fig, ax = plt.subplots(figsize=(14, 14))

    bar_height = 0.3
    assist_height = 0.15
    bar_offset = 3.5
    circle_radius = 0.6

    for i, y in enumerate(y_positions):
        year = all_years[i]

        # Messi side (light blue background)
        ax.barh(y=y + 0.25, width=-messi_stats.loc[i, 'goals'], height=bar_height,
                label='Goals' if i == 0 else "", color='#0057B7', left=-bar_offset)  # Royal Blue
        ax.barh(y=y - 0.25, width=-messi_stats.loc[i, 'assists'], height=assist_height,
                label='Assists' if i == 0 else "", color='#00C2CB', left=-bar_offset)  # Aqua

        # Ronaldo side (dark red background)
        ax.barh(y=y + 0.25, width=ronaldo_stats.loc[i, 'goals'], height=bar_height,
                color='#8B0000', left=bar_offset)  # Dark Red
        ax.barh(y=y - 0.25, width=ronaldo_stats.loc[i, 'assists'], height=assist_height,
                color='#FFB84D', left=bar_offset)  # Warm Gold

        # Center circle + year text
        circle = plt.Circle((0, y), circle_radius, color='black', zorder=5)
        ax.add_patch(circle)
        ax.text(0, y, str(year), color='white', ha='center', va='center',
                fontsize=9, fontweight='bold', zorder=6)

    # Flip vertical order
    ax.set_ylim(max(y_positions) + spacing, -spacing)

    # Show x-axis, hide y-axis
    ax.set_yticks([])
    ax.set_yticklabels([])
    ax.tick_params(axis='y', left=False)
    ax.spines['left'].set_visible(False)

    ax.grid(axis='x', linestyle='--', alpha=0.4)
    ax.set_xlabel('Goals & Assists', fontsize=12)
    ax.set_title('Messi vs Ronaldo: Yearly Goals & Assists', fontsize=16, pad=20)
    ax.legend(loc='upper left')
    plt.tight_layout()
    plt.show()
```

```python
def plot_clean_equal_bar_heights_goals_assists_svg(messi_df, ronaldo_df):
    import pandas as pd
    import matplotlib.pyplot as plt

    # Convert to year
    messi_df['year'] = pd.to_datetime(messi_df['date']).dt.year.astype(int)
    ronaldo_df['year'] = pd.to_datetime(ronaldo_df['date']).dt.year.astype(int)

    # Filter for 2009â€“2017
    messi_df = messi_df[(messi_df['year'] >= 2009) & (messi_df['year'] <= 2017)]
    ronaldo_df = ronaldo_df[(ronaldo_df['year'] >= 2009) & (ronaldo_df['year'] <= 2017)]

    # Group by year
    messi_stats = messi_df.groupby('year')[['goals', 'assists']].sum().reset_index()
    ronaldo_stats = ronaldo_df.groupby('year')[['goals', 'assists']].sum().reset_index()

    all_years = list(range(2009, 2018))
    spacing = 2.2
    y_positions = [i * spacing for i in range(len(all_years))]

    messi_stats = messi_stats.set_index('year').reindex(all_years, fill_value=0).reset_index()
    ronaldo_stats = ronaldo_stats.set_index('year').reindex(all_years, fill_value=0).reset_index()

    fig, ax = plt.subplots(figsize=(14, 14), facecolor='none')

    bar_height = 0.3
    bar_offset = 3.5

    GOAL_COLOR = '#F4D35E'
    ASSIST_COLOR = '#1B998B'

    for i, y in enumerate(y_positions):
        # Messi Goals
        g_val = -messi_stats.loc[i, 'goals']
        ax.barh(y=y + 0.25, width=g_val, height=bar_height, color=GOAL_COLOR, left=-bar_offset)
        ax.text(g_val - bar_offset, y + 0.25, str(int(messi_stats.loc[i, 'goals'])),
                ha='right', va='center', fontsize=10)

        # Messi Assists
        a_val = -messi_stats.loc[i, 'assists']
        ax.barh(y=y - 0.25, width=a_val, height=bar_height, color=ASSIST_COLOR, left=-bar_offset)
        ax.text(a_val - bar_offset, y - 0.25, str(int(messi_stats.loc[i, 'assists'])),
                ha='right', va='center', fontsize=10)

        # Ronaldo Goals
        g_val_r = ronaldo_stats.loc[i, 'goals']
        ax.barh(y=y + 0.25, width=g_val_r, height=bar_height, color=GOAL_COLOR, left=bar_offset)
        ax.text(g_val_r + bar_offset, y + 0.25, str(int(ronaldo_stats.loc[i, 'goals'])),
                ha='left', va='center', fontsize=10)

        # Ronaldo Assists
        a_val_r = ronaldo_stats.loc[i, 'assists']
        ax.barh(y=y - 0.25, width=a_val_r, height=bar_height, color=ASSIST_COLOR, left=bar_offset)
        ax.text(a_val_r + bar_offset, y - 0.25, str(int(ronaldo_stats.loc[i, 'assists'])),
                ha='left', va='center', fontsize=10)

    # Clean layout
    ax.set_ylim(max(y_positions) + spacing, -spacing)
    ax.set_yticks([])
    ax.set_yticklabels([])
    ax.tick_params(axis='both', left=False, bottom=False, labelleft=False, labelbottom=False)
    for spine in ax.spines.values():
        spine.set_visible(False)
    ax.set_facecolor('none')

    plt.tight_layout()
    plt.savefig("graph1.svg", format='svg')
    plt.close()
    return "graph1.svg"
```

```python
def plot_clean_key_passes_over_time_svg(messi_df, ronaldo_df):
    # Convert to year
    messi_df['year'] = pd.to_datetime(messi_df['date']).dt.year.astype(int)
    ronaldo_df['year'] = pd.to_datetime(ronaldo_df['date']).dt.year.astype(int)

    # Filter: 2009â€“2017
    messi_df = messi_df[(messi_df['year'] >= 2009) & (messi_df['year'] <= 2017)]
    ronaldo_df = ronaldo_df[(ronaldo_df['year'] >= 2009) & (ronaldo_df['year'] <= 2017)]

    # Group key passes
    messi_kp = messi_df.groupby('year')['key_passes'].sum().reset_index()
    ronaldo_kp = ronaldo_df.groupby('year')['key_passes'].sum().reset_index()

    # Define consistent years and spacing
    all_years = list(range(2009, 2018))
    spacing = 2.2
    y_positions = [i * spacing for i in range(len(all_years))]

    messi_kp = messi_kp.set_index('year').reindex(all_years, fill_value=0).reset_index()
    ronaldo_kp = ronaldo_kp.set_index('year').reindex(all_years, fill_value=0).reset_index()

    # Plot setup
    fig, ax = plt.subplots(figsize=(14, 14), facecolor='none')
    bar_height = 0.3
    bar_offset = 3.5
    text_padding = 1.0
    BAR_COLOR = '#F4D35E'  # Golden Yellow

    for i, y in enumerate(y_positions):
        # Messi Bar (left)
        val_m = -messi_kp.loc[i, 'key_passes']
        ax.barh(y=y, width=val_m, height=bar_height, color=BAR_COLOR, left=-bar_offset)
        ax.text(val_m - bar_offset - text_padding, y, str(int(messi_kp.loc[i, 'key_passes'])),
                ha='right', va='center', fontsize=10)

        # Ronaldo Bar (right)
        val_r = ronaldo_kp.loc[i, 'key_passes']
        ax.barh(y=y, width=val_r, height=bar_height, color=BAR_COLOR, left=bar_offset)
        ax.text(val_r + bar_offset + text_padding, y, str(int(ronaldo_kp.loc[i, 'key_passes'])),
                ha='left', va='center', fontsize=10)

    # Clean visual layout
    ax.set_ylim(max(y_positions) + spacing, -spacing)
    ax.set_yticks([])
    ax.set_yticklabels([])
    ax.tick_params(axis='both', left=False, bottom=False, labelleft=False, labelbottom=False)
    for spine in ax.spines.values():
        spine.set_visible(False)
    ax.set_facecolor('none')

    plt.tight_layout()
    plt.savefig("graph2.svg", format='svg')
    plt.close()

    return "graph2.svg"
```

```python
def plot_clean_total_goals_assists_by_stage_svg(messi_df, ronaldo_df):
    # Extract year
    messi_df['year'] = pd.to_datetime(messi_df['date']).dt.year.astype(int)
    ronaldo_df['year'] = pd.to_datetime(ronaldo_df['date']).dt.year.astype(int)

    # Filter 2009â€“2017
    messi_df = messi_df[(messi_df['year'] >= 2009) & (messi_df['year'] <= 2017)]
    ronaldo_df = ronaldo_df[(ronaldo_df['year'] >= 2009) & (ronaldo_df['year'] <= 2017)]

    # Knockout stages
    stages = ['Final', 'SF', 'QF', 'R16']
    spacing = 2.2
    y_positions = [i * spacing for i in range(len(stages))]

    messi_goals, messi_assists = [], []
    ronaldo_goals, ronaldo_assists = [], []

    for stage in stages:
        m = messi_df[messi_df['stage'] == stage]
        r = ronaldo_df[ronaldo_df['stage'] == stage]
        messi_goals.append(m['goals'].sum())
        messi_assists.append(m['assists'].sum())
        ronaldo_goals.append(r['goals'].sum())
        ronaldo_assists.append(r['assists'].sum())

    # Adjust figure size to match the 9-bar visuals
    fig_height = 14 * (len(stages) / 9)
    fig, ax = plt.subplots(figsize=(14, fig_height), facecolor='none')

    bar_height = 0.3
    bar_offset = 3.5
    text_padding = 1.0
    GOAL_COLOR = '#F4D35E'
    ASSIST_COLOR = '#1B998B'

    for i, y in enumerate(y_positions):
        # Messi (left)
        g_val = -messi_goals[i]
        a_val = -messi_assists[i]
        ax.barh(y=y + 0.25, width=g_val, height=bar_height,
                color=GOAL_COLOR, left=-bar_offset)
        ax.text(g_val - bar_offset - text_padding, y + 0.25, str(int(messi_goals[i])),
                ha='right', va='center', fontsize=10)

        ax.barh(y=y - 0.25, width=a_val, height=bar_height,
                color=ASSIST_COLOR, left=-bar_offset)
        ax.text(a_val - bar_offset - text_padding, y - 0.25, str(int(messi_assists[i])),
                ha='right', va='center', fontsize=10)

        # Ronaldo (right)
        g_val_r = ronaldo_goals[i]
        a_val_r = ronaldo_assists[i]
        ax.barh(y=y + 0.25, width=g_val_r, height=bar_height,
                color=GOAL_COLOR, left=bar_offset)
        ax.text(g_val_r + bar_offset + text_padding, y + 0.25, str(int(ronaldo_goals[i])),
                ha='left', va='center', fontsize=10)

        ax.barh(y=y - 0.25, width=a_val_r, height=bar_height,
                color=ASSIST_COLOR, left=bar_offset)
        ax.text(a_val_r + bar_offset + text_padding, y - 0.25, str(int(ronaldo_assists[i])),
                ha='left', va='center', fontsize=10)

    ax.set_ylim(max(y_positions) + spacing, -spacing)
    ax.set_yticks([])
    ax.set_yticklabels([])
    ax.tick_params(axis='both', left=False, bottom=False, labelleft=False, labelbottom=False)
    for spine in ax.spines.values():
        spine.set_visible(False)
    ax.set_facecolor('none')

    plt.tight_layout()
    plt.savefig("graph3.svg", format='svg')
    plt.close()

    return "graph3.svg"
plot_clean_total_goals_assists_by_stage_svg(messi_df, ronaldo_df)
```
